```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
install_df = pd.read_csv('installments_payments.csv')
```

```python
install_df.tail(20)
```

```python
install_df.shape
```

```python
curr_app = pd.read_csv('application_train.csv')
y = curr_app['TARGET']
X = curr_app.iloc[:, :1]
```

```python
X_test = pd.read_csv('application_test.csv')
X_test = X_test.iloc[:, :1]
```

```python
install_df['DAYS_LATE_PAYMENT'] = install_df.DAYS_INSTALMENT-install_df.DAYS_ENTRY_PAYMENT
install_df['AMT_SHORT_PAYMENT'] = install_df.AMT_INSTALMENT-install_df.AMT_PAYMENT
```

```python
install_df.head()
```

```python
gb_sum_df = install_df[['AMT_INSTALMENT', 'AMT_PAYMENT', 'DAYS_LATE_PAYMENT', 'AMT_SHORT_PAYMENT', 'SK_ID_CURR']].groupby('SK_ID_CURR').sum().reset_index()
```

```python
gb_sum_df['COUNT'] = install_df[['SK_ID_CURR']].groupby('SK_ID_CURR').size().reset_index().iloc[:, 1]
```

```python
gb_sum_df['AVG_DAYS_LATE'] = gb_sum_df.DAYS_LATE_PAYMENT / gb_sum_df.COUNT
gb_sum_df['AVG_AMT_SHORT'] = gb_sum_df.AMT_SHORT_PAYMENT / gb_sum_df.AMT_INSTALMENT
gb_sum_df['AVG_AMT_SHORT'] = gb_sum_df['AVG_AMT_SHORT'].replace(-np.inf, np.nan)
```

```python
gb_sum_df.head()
```

```python
gb_sum_df['MISSED_PAYMENTS'] = pd.concat([install_df[['AMT_PAYMENT']].isnull(), install_df[['SK_ID_CURR']]], axis=1).groupby('SK_ID_CURR').sum().reset_index().iloc[:, 1]
```

```python
gb_sum_df[gb_sum_df.MISSED_PAYMENTS > 0]
```

```python
X = X.merge(gb_sum_df, on='SK_ID_CURR', how='left')
X_test = X_test.merge(gb_sum_df, on='SK_ID_CURR', how='left')
```

```python
corrs = pd.Series([y.corr(X[col]) for col in X.columns[-gb_sum_df.shape[1]+1:]], index=gb_sum_df.columns[1:]).sort_values()
```

```python
corrs = abs(corrs).sort_values(ascending=False)
corrs
```

```python
X[corrs.index].to_pickle('features_install.pkl')
X_test[corrs.index].to_pickle('features_install_test.pkl')
```
