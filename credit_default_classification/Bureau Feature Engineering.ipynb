```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
curr_app = pd.read_csv('application_train.csv')
```

```python
y = curr_app['TARGET']
X = curr_app.iloc[:, :1]
```

```python
X_test = pd.read_csv('application_test.csv')
X_test = X_test.iloc[:, :1]
```

```python
bureau_df = pd.read_csv('bureau.csv')
```

```python
[print(col, bureau_df[col].value_counts()) for col in bureau_df.select_dtypes(include=object)]
```

```python
bureau_balance_df = pd.read_csv('bureau_balance.csv')
```

```python
bureau_balance_df.shape
```

```python
bureau_balance_df['STATUS'] = bureau_balance_df['STATUS'].replace(['C', 'X'], 0).astype(int)
```

```python
gb_max_status_bureau = bureau_balance_df[['SK_ID_BUREAU', 'STATUS']].groupby('SK_ID_BUREAU').max().reset_index()
```

```python
bureau_df.head()
```

```python
# bureau_df['TERM'] = bureau_df['AMT_CREDIT_SUM'] / bureau_df['AMT_ANNUITY']
```

```python
# bureau_df['TERM'] = bureau_df['TERM'].replace(np.inf, np.nan)
```

```python
bureau_df = bureau_df.merge(gb_max_status_bureau, how='left', on='SK_ID_BUREAU')
```

```python
bureau_df.head()
```

```python
groupby_max_df = bureau_df[['SK_ID_CURR', 'STATUS']].groupby('SK_ID_CURR').max()
groupby_max_df.head()
```

```python
# groupby_sum_df = (bureau_df[['SK_ID_CURR', 'CREDIT_DAY_OVERDUE', 
#                            'AMT_CREDIT_MAX_OVERDUE', 'AMT_CREDIT_SUM_OVERDUE',
#                            'AMT_CREDIT_SUM', 'AMT_CREDIT_SUM_DEBT', 'AMT_ANNUITY']]
#                 .groupby('SK_ID_CURR')
#                 .sum().reset_index()
#                )
# groupby_sum_df.head()
```

```python
bureau_df = pd.get_dummies(bureau_df)
```

```python
bureau_df.head()
```

```python
gb_sum_df = bureau_df.groupby('SK_ID_CURR').sum().reset_index().drop(columns=['SK_ID_BUREAU'])
```

```python
# groupby_mean_df = (bureau_df[['SK_ID_CURR', 'TERM']]
#                 .groupby('SK_ID_CURR')
#                 .mean().reset_index()
#                )
# groupby_mean_df['TERM'].mean()
```

```python
gb_sum_df['TERM'] = gb_sum_df['AMT_CREDIT_SUM'] / gb_sum_df['AMT_ANNUITY']
```

```python
gb_sum_df['TERM'] = gb_sum_df['TERM'].replace(np.inf, np.nan)
```

```python
gb_sum_df['STATUS'] = groupby_max_df['STATUS']
```

```python
gb_sum_df.head()
```

```python
X = X.merge(gb_sum_df, on='SK_ID_CURR', how='left')
X_test = X_test.merge(gb_sum_df, on='SK_ID_CURR', how='left')
```

```python
X.head()
```

```python
corrs = pd.Series([y.corr(X[col]) for col in X.columns[-gb_sum_df.shape[1]+1:]], index=gb_sum_df.columns[1:]).sort_values()
```

```python
corrs = abs(corrs).sort_values(ascending=False)
corrs
```

```python
X[corrs.index].to_pickle('features_bureau.pkl')
X_test[corrs.index].to_pickle('features_bureau_test.pkl')
```
