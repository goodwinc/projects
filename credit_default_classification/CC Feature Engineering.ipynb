```python
curr_app = pd.read_csv('application_train.csv')
y = curr_app['TARGET']
X = curr_app.iloc[:, :1]
```

```python
X_test = pd.read_csv('application_test.csv')
X_test = X_test.iloc[:, :1]
```

```python
cc_df = pd.get_dummies(cc_df)
```

```python
cc_df['PERCENT_DRAW_LIMIT'] = (cc_df['AMT_DRAWINGS_CURRENT'] / cc_df['AMT_CREDIT_LIMIT_ACTUAL']).replace([-np.inf, np.inf], np.nan)
cc_df['PERCENT_DRAW_ATM_LIMIT'] = (cc_df['AMT_DRAWINGS_ATM_CURRENT'] / cc_df['AMT_CREDIT_LIMIT_ACTUAL']).replace([-np.inf, np.inf], np.nan)
```

```python
cc_df.PERCENT_DRAW_ATM_LIMIT.describe()
```

```python
gb_sum_df = cc_df.iloc[:, 1:].groupby('SK_ID_CURR').sum().reset_index()
```

```python
gb_sum_df['COUNT'] = cc_df[['SK_ID_CURR']].groupby('SK_ID_CURR').size().reset_index().iloc[:, 1]
```

```python
gb_sum_df['CNT_DRAWINGS_CURRENT_MAX'] = cc_df[['CNT_DRAWINGS_CURRENT', 'SK_ID_CURR']].groupby('SK_ID_CURR').max().reset_index().iloc[:, 1]
gb_sum_df['CNT_DRAWINGS_CURRENT_AVG'] = cc_df[['CNT_DRAWINGS_CURRENT', 'SK_ID_CURR']].groupby('SK_ID_CURR').mean().reset_index().iloc[:, 1]

gb_sum_df['CNT_DRAWINGS_ATM_CURRENT_MAX'] = cc_df[['CNT_DRAWINGS_ATM_CURRENT', 'SK_ID_CURR']].groupby('SK_ID_CURR').max().reset_index().iloc[:, 1]
gb_sum_df['CNT_DRAWINGS_ATM_CURRENT_AVG'] = cc_df[['CNT_DRAWINGS_ATM_CURRENT', 'SK_ID_CURR']].groupby('SK_ID_CURR').mean().reset_index().iloc[:, 1]

gb_sum_df['AMT_CREDIT_LIMIT_ACTUAL_AVG'] = cc_df[['AMT_CREDIT_LIMIT_ACTUAL', 'SK_ID_CURR']].groupby('SK_ID_CURR').mean().reset_index().iloc[:, 1]
```

```python
gb_sum_df['SK_DPD_AVG'] = (cc_df[['SK_ID_CURR','SK_DPD']]
 .groupby(['SK_ID_CURR']).mean().reset_index().iloc[:,1]
)

gb_sum_df['SK_DPD_DEF_AVG'] = (cc_df[['SK_ID_CURR','SK_DPD_DEF']]
 .groupby(['SK_ID_CURR']).mean().reset_index().iloc[:,1]
)
```

```python
gb_sum_df['PERCENT_DRAW_LIMIT'] = cc_df[['PERCENT_DRAW_LIMIT', 'SK_ID_CURR']].groupby('SK_ID_CURR').mean().reset_index().iloc[:, 1]
gb_sum_df['PERCENT_DRAW_ATM_LIMIT'] = cc_df[['PERCENT_DRAW_ATM_LIMIT', 'SK_ID_CURR']].groupby('SK_ID_CURR').mean().reset_index().iloc[:, 1]
```

```python
X = X.merge(gb_sum_df, on='SK_ID_CURR', how='left')
```

```python
X_test = X_test.merge(gb_sum_df, on='SK_ID_CURR', how='left')
```

```python
corrs = pd.Series([y.corr(X[col]) for col in X.columns[-gb_sum_df.shape[1]+1:]], index=gb_sum_df.columns[1:]).sort_values()
```

```python
corrs = abs(corrs).sort_values(ascending=False)
corrs
```
